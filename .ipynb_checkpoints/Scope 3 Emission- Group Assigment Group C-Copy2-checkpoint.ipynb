## Conversion DF Evaluation

```python
df_raw = pd.read_excel('df_evaluation4.xlsx',sheet_name=0)
```

Inbound outbound
Remove legs
Only Europe
Add Sites

```python
df_con=df_raw
#Inbound out bound
df_con['source']=np.where(df_con ['source']=='GCR' ,'Outbound','Inbound')
#Only Europe
df_EU=df_con[(df_con['Origin_Region']=='Europe') | (df_con['Target_Region']=='Europe')]
#Only Road
df_EU=df_EU[df_EU['leg2_mot']!='Vessel']
#Clean up
df_EU=df_EU[(df_EU['Distance']>0)&(df_EU['Origin_Postal_Code']!='.') & (df_EU['Target_Postal_Code']!='.')]
#Only selected columns
df_clean=df_EU[['Origin_Postal_Code', 'Origin_City',
       'Origin_Country', 'Target_Postal_Code', 'Target_City', 'Target_Country',
       'Amount', 'source', 'Origin_Country_Name', 'Origin_Region',
       'Target_Country_Name', 'Target_Region', 'Origin_Key',
       'Target_Key', 'Distance']]
```

```python
#Calculation of Sites
df_sites = pd.read_excel('PbP Download.xlsx',sheet_name=1)
df_sites2 = pd.read_excel('PbP Download.xlsx',sheet_name=2)
```

```python
#Sites inbound
df_sites=df_sites[[ 'Org_Plant SAP',  'Postal Code', 'City','Country']]
df_sites=df_sites.drop_duplicates(subset='Org_Plant SAP', keep="last")
df_sites.columns=['Plant','Postal_Code','City','Country']
#Sites Outbound
df_sites2=df_sites2[[ 'G0.Plant',  'Postal Code', 'City','Country']]
df_sites2=df_sites2.drop_duplicates(subset='G0.Plant', keep="last")
df_sites2.columns=['Plant','Postal_Code','City','Country']
#Union of the 2
df_sites3=pd.concat([df_sites,df_sites2], axis=0)
df_sites3['Key']=df_sites3['Postal_Code']+'-'+df_sites3['Country']
df_sites3=df_sites3.drop_duplicates(subset='Key', keep="last")
```

```python
#Merge Data with Sites
df_clean_enh= pd.merge(df_clean, df_sites3[['Key','Plant']], left_on="Origin_Key",right_on="Key", how="left")
df_clean_enh= pd.merge(df_clean_enh, df_sites3[['Key','Plant']], left_on="Target_Key",right_on="Key", how="left")
# df_clean_enh.columns
df_clean_enh.drop(['Key_x','Key_y'], axis=1,inplace=True)
df_clean_enh=df_clean_enh.reindex(['Origin_Key','Origin_Postal_Code', 'Origin_City', 'Origin_Country', 'Origin_Country_Name','Origin_Region','Plant_x',
              'Target_Key','Target_Postal_Code', 'Target_City', 'Target_Country','Target_Country_Name','Target_Region','Plant_y',
               'source',  'Amount','Distance'],axis="columns")
df_clean_enh2=df_clean_enh.rename(columns={'Plant_x': "Origin_Plant", 'Plant_y': "Target_Plant", 'source': "InBound-OutBound"})
df_clean_enh2['C02 Emission']=(df_clean_enh2['Amount']*df_clean_enh2['Distance'])/1000
df_clean_enh2
```

```python
# df_clean_enh2.to_excel('df_clean_enh2.xlsx')
```

## Data Cleaning

```python
# df_raw = pd.read_excel('Data sample.xlsx',sheet_name=0)
df_raw=df_clean_enh2
```

```python
df=df_raw[(df_raw['Distance']>200) & (df_raw['Origin_Postal_Code']!='.')& (df_raw['Target_Postal_Code']!='.')]
df.dropna(subset='Origin_Postal_Code',inplace=True)
df.dropna(subset='Target_Postal_Code',inplace=True)
df=df.head(300)
# df.to_excel('Data sample.xlsx')
```

## Preparation of Dataframes to create a Graph

```python
df_origin=df[['Origin_Key','Origin_Postal_Code','Origin_City','Origin_Country','Origin_Region']]
df_origin.columns=['Key','Postal_Code','City','Country','Region']
#df_origin
df_Target=df[['Target_Key','Target_Postal_Code','Target_City','Target_Country','Target_Region']]
df_Target.columns=['Key','Postal_Code','City','Country','Region']
# df_Target
df_locations=pd.concat([df_origin,df_Target], axis=0)
df_locations=df_locations.dropna()
# df_locations
df_locations_unique=df_locations.drop_duplicates(subset='Key', keep="last")
# df_locations_unique
#df_flows=df[['Origin_Key','Target_Key','Distance']] I would like to use distance as wiehgt but it is looking bad on the chart. Need to research
df_flows=df[['Origin_Key','Target_Key']]
# df_flows
```

## Creation of the Graph

```python
# pip install pyvis
```

```python
from pyvis.network import Network
net = Network(notebook=True)
#net.from_nx(graph)
#net.show_buttons()
#network.Network.set_options()
#net.show("example.html")
```

#### Adding nodes and edges 

```python
nodes = df_locations_unique['Key'].to_list()
nodes = [node.strip() for node in nodes]
net.add_nodes(nodes) 
edges=df_flows.values.tolist()
net.add_edges(edges)
net.repulsion(100, spring_length=500)
net.show_buttons(filter_=['physics'])
net.show("match.html")
```
